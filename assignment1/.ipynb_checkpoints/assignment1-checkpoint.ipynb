### (b) Visualize the order of feature importance.
#### feature importance: The chi-square test helps you to solve the problem in feature selection by testing the relationship between the features. When two features are independent, the observed count is close to the expected count, thus we will have smaller Chi-Square value. So high Chi-Square value indicates that the hypothesis of independence is incorrect. In simple words, higher the Chi-Square value the feature is more dependent on the response and it can be selected for model training. However, the cons is Chi-Square is sensitive to small frequencies in cells of tables. Generally, when the expected value in a cell of a table is less than 5, chi-square can lead to errors in conclusions. Luckily, we have large training dataset. 




```python
chi2_score=chi2(testDF,targetDF)
```

# 3. Select features 
### (based on Chi-Square)
#### when selected features, we select the feature mannually. that's what we did in the 1. clean data. After that, we use chi-square to select the top 30 ranked feature.

#### The chi-square test helps you to solve the problem in feature selection by testing the relationship between the features. When two features are independent, the observed count is close to the expected count, thus we will have smaller Chi-Square value. So high Chi-Square value indicates that the hypothesis of independence is incorrect. In simple words, higher the Chi-Square value the feature is more dependent on the response and it can be selected for model training. 
#### The top 30 features are the more dependent on our target, therefore, it makes sense that we use this to train model

### 4.2 build ordinal multi-class classification (ordinal logistic regression)

```python

def multi_class_ordinary_train(X_train, y_train,X_test,param,solver):
    model = LogisticRegression(C=param,solver=solver) 
    probability_train=[]
    prob_0=0
    for i in range(14):
        y_tem=list(map(lambda x: 0 if x<=i else 1,y_train))
        if i ==13:
            model.fit(X_train, y_tem)
            prob_a = model.predict_proba(X_train)[:,0].reshape(len(X_train),1)
            prob_b = model.predict_proba(X_train)[:,1].reshape(len(X_train),1)
            prob_1=prob_a-prob_0
            probability_train.append(prob_1)
            probability_train.append(prob_b)
        else:
            model.fit(X_train, y_tem)
            prob_a = (model.predict_proba(X_train)[:,0]).reshape(len(X_train),1)
            prob_1=prob_a-prob_0
            probability_train.append(prob_1)
            prob_0=prob_a 
            
    probability_train=np.array(probability_train)

    #pro=pro.reshape(15,1054)
    probability_train=probability_train.reshape(15,len(X_train))
    probability_train=np.transpose(probability_train)
    #probability=probability.reshape(int(len(probability)/15),15)
    return probability_train


def multi_class_ordinary_test(X_train, y_train,X_test,param,solver):
    model = LogisticRegression(C=param,solver=solver) 
    probability_test=[]
    prob_0=0
    for i in range(14):
        y_tem=list(map(lambda x: 0 if x<=i else 1,y_train))
        if i ==13:
            model.fit(X_train, y_tem)
            prob_a = model.predict_proba(X_test)[:,0].reshape(len(X_test),1)
            prob_b = model.predict_proba(X_test)[:,1].reshape(len(X_test),1)
            prob_1=prob_a-prob_0
            probability_test.append(prob_1)
            probability_test.append(prob_b)
        else:
            model.fit(X_train, y_tem)
            prob_a = (model.predict_proba(X_test)[:,0]).reshape(len(X_test),1)
            prob_1=prob_a-prob_0
            probability_test.append(prob_1)
            prob_0=prob_a 
            
    probability_test=np.array(probability_test)

    #pro=pro.reshape(15,1054)
    probability_test=probability_test.reshape(15,len(X_test))
    probability_test=np.transpose(probability_test)
    #probability=probability.reshape(int(len(probability)/15),15)
    return probability_test
```

### 5.1 grid search

#### penalty{‘l1’, ‘l2’, ‘elasticnet’, ‘none’}, default=’l2’ Used to specify the norm used in the penalization. The ‘newton-cg’, ‘sag’ and ‘lbfgs’ solvers support only l2 penalties. ‘elasticnet’ is only supported by the ‘saga’ solver. If ‘none’ (not supported by the liblinear solver), no regularization is applied. Here, the logistic regression model is binary logistic regression. Therefore, we have to use l2 to hyperparameter tuning, so that we can use solver'sag' and 'lbfgs'

#### In this result, we use accuracy as our metric to measure the model performance. The result shows that the optimal log model uses C=0.1, and a newton-cg solver, and has a cross validation score of 31.924% with a standard deviation of 0.387%
